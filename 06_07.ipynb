## 네이버블로그_자동스크롤내리기

```python
from selenium import webdriver  as wb
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
import pandas as pd
from bs4 import BeautifulSoup as bs
```

```python
driver = wb.Edge()
driver.get('http://www.naver.com')
time.sleep(2)
```

```python
#1. id가  query인곳에 키워드를 "인공지능" 입력하고 클릭
search=driver.find_element(By.ID,'query')
search.send_keys("인공지능")
search.submit()  #click()과 같음 
time.sleep(1)
```

```python
#2.VIEW 메뉴를 클릭
# //*[@id="lnb"]/div[1]/div/ul/li[6]/a
driver.find_element(By.XPATH,'//*[@id="lnb"]/div[1]/div/ul/li[6]/a').click()
```

```python
#3.VIEW 메뉴 중에서 '전체'에서 '블로그'클릭 
# //*[@id="snb"]/div[1]/div/div[1]/a[2]
driver.find_element(By.XPATH,'//*[@id="snb"]/div[1]/div/div[1]/a[2]').click()
```

```python
#4.옵션 클릭 
# //*[@id="snb"]/div[1]/div/div[2]/a
driver.find_element(By.XPATH,'//*[@id="snb"]/div[1]/div/div[2]/a').click()
```

```python
#5.정렬에 최신순 클릭 
# //*[@id="snb"]/div[2]/ul/li[2]/div/div/a[2]
driver.find_element(By.XPATH,'//*[@id="snb"]/div[2]/ul/li[2]/div/div/a[2]').click()
```

### 게시글 제목, 내용, 작성자, 작성일


```python
html = driver.page_source
soup = bs(html, 'html.parser')
```

```python
# 게시글 제목
titles = soup.find_all('a', 'api_txt_lines total_tit')
len(titles)
```

```python
# 게시글 내용
contents = soup.find_all('div', 'api_txt_lines dsc_txt')
len(contents)
```

```python
# 게시글 작성자 
names = soup.find_all('a', 'sub_txt sub_name')
len(names)
```

```python
# 게시글 작성일
date = soup.find_all('span', 'sub_time sub_txt')
len(date)
```

```python
num_list = []
names_list = []
titles_list = []
contents_list = []
names_list = []
dates_list = []

blog_list=soup.find_all('div', 'total_wrap api_ani_send')
print('========== blog_list' + "=" * 60)

num=1 #게시글번호
f = open('인공지능.txt', 'w', encoding='UTF-8')

for i in blog_list:
    #1. 게시글 번호
    num_list.append(num)
    f.write("번호 : "+str(num)+'\n') #txt에 번호 쓰기
    print(f"번호:{num}"+'\n')
    
    #2. 게시글 제목
    title=i.find('a','api_txt_lines total_tit').text
    titles_list.append(title)
    f.write("제목 :"+str(title)+'\n')
    print(f"제목 : {title}"+'\n')
    
    #3. 게시글 내용
    content = i.find('div', 'api_txt_lines dsc_txt').text
    contents_list.append(content)
    f.write("내용 :"+str(content)+'\n')
    print(f"내용 : {content}"+'\n')
    
    #4. 게시글 작성자
    name = i.find('a', 'sub_txt sub_name').text
    names_list.append(name)
    f.write("작성자 :"+str(name)+'\n')
    print(f"작성자 : {name}"+'\n')
    
    #4. 게시글 작성일
    date = i.find('span', 'sub_time sub_txt').text
    dates_list.append(date)
    f.write("작성일 :"+str(date)+'\n')
    print(f"작성일 : {date}"+'\n')
    
    print("="*80)
    num = num+1

f.close()    
```

```python
import pandas as pd

#딕셔너리 만들기

df=pd.DataFrame({'번호' : num_list, '제목': titles_list, '내용': contents_list, '작성자': names_list, '작성일': dates_list})
df.head(5)
```
