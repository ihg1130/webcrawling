```python
import requests as req
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
url = "https://pjt3591oo.github.io/"
res = req.get(url)
soup = bs(res.content, 'lxml')
```

```python
#body > main > div > div > div:nth-child(9) > h3 > a
titles = soup.select('div > h3 > a')
len(titles)
```

```python
for i in titles:
    title = i.text
    print(title)
```

```python
contents = soup.select('h4')
for i in contents :
    content = i.text.strip()
    print(content)
```

```python
#body > main > div > div > div:nth-child(9) > p > span
names = soup.find_all('span','date')
for i in names:
    print(i.text[-3:])
```

```python
#https://pjt3591oo.github.io/page3/
num=2
while True:
    sub_url = 'page' + str(num)
    path = url + sub_url
    res = req.get(path)
    soup = bs(res.content, 'lxml')
    titles=soup.select('div > h3 > a')
    contents = soup.select('h4')
    authors = []
    
    for i in names:
        authors.append(i.text[-3:])
        
    for i in range(len(titles)):
        print("제목 : " + titles[i].text)
        print("내용 : " + contents[i].text.strip())
        print("저자 : " + authors[i])
        print("-"*50)
        
    num +=1
    if res.status_code != 200:
        print("종료" + str(res.status_code))
        break
```

```python
#https://movie.daum.net/ranking/boxoffice/yearly
url = "https://movie.daum.net/ranking/boxoffice/yearly"
res = req.get(url)
soup = bs(res.content, 'lxml')
```

```python
#영화제목
movies = soup.select('div > div.box_boxoffice > ol >li')
for m in movies:
    title = m.select_one('div.thumb_cont > strong > a').text
    print(title)
```

```python
#영화 줄거리
movies = soup.select('div > div.box_boxoffice > ol >li')
for m in movies:
    story = m.select_one('div.poster_info > a').text.strip()
    print(story)
```

```python
#영화 제목
movies = soup.select('div > div.box_boxoffice > ol >li')
for m in movies :
    img_url = m.select_one('div.poster_movie > img')['src']
    print(img_url)
```

```python
movies = soup.select('div > div.box_boxoffice > ol >li')
for m in movies:
    title = m.select_one('div.thumb_cont > strong > a').text
    story = m.select_one('div.poster_info > a').text.strip()   
    img_url = m.select_one('div.poster_movie > img')['src']
    print("영화제목 : " + title)
    print("\n")
    print("줄거리 : " + story)
    print("\n")
    print("이미지 : " + img_url)
    print("\n")
```

### 영화 포스터 이미지 저장하기

```python
url = "https://movie.daum.net/ranking/boxoffice/yearly"
res = req.get(url)
soup = bs(res.content, 'lxml')
movies = soup.select('div.box_boxoffice > ol > li')
for idx,m in enumerate(movies):
    img_url = m.select_one('div.poster_movie > img')['src']
    #print(idx, img_url)
    img_res = req.get(img_url)
    
    with open('img/movie_{}.jpg'.format(idx+1), 'wb') as f:
        f.write(img_res.content)
```
