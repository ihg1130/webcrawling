```python
res = req.get("http://www.yes24.com/24/Category/BestSeller")
soup = bs(res.content, "lxml")
```

```python
#bestList > ol > li.num1 > p:nth-child(3) > a -> 책제목

#책 설명만 출력됨
books = soup.select("p.copy > a")
for i in books:
    print(i.text)
```
