#### selenium 라이브러리
- 웹 브라우저를 자동으로 제어하기 위한 라이브러리
- 브라우저를 뛰우는것부터 제어 -> request가 필요 없음
- wevdriver : 브라우저 역할(request 역할)
- Keys : 컴퓨터에서 키보드 역할
- By : 선택자를 선택하는 역할

```python
from selenium import webdriver  as wb
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
import pandas as pd
from bs4 import BeautifulSoup as bs
```

```python
#1. 크롬창 열기
driver = wb.Chrome()
```

```python
#2. 네이버 접속하기 : driver.get(url)
driver.get("https://www.naver.com")
```

```python
#3. 검색바 찾기 id : query
search = driver.find_element(By.ID,'query')
```

```python
#4. 키워드 입력
search.send_keys('광주날씨')
```

```python
#5.검색하기
#5-1. 엔터키 누르기
search.send_keys(Keys.ENTER)
```

```python
#6. 뒤로가기
driver.back()
```

```python
#7. 창종료하기
driver.close()
```

```python
#검색 엔터
driver = wb.Chrome()
driver.get("https://www.naver.com")
search = driver.find_element(By.ID,'query')
# == driver.find_element_by_id('query')
search.send_keys('웹크롤링')
search.send_keys(Keys.ENTER)
```

```python
#검색 버튼 태그 요소
btn_search = driver.find_element(By.ID,'search_btn')
```

```python
#검색버튼 클릭
btn_search.click()
```

```python
#검색버튼 클릭으로 검색
driver = wb.Chrome()
driver.get("https://www.naver.com")
btn_search = driver.find_element(By.ID,'search_btn')
btn_search.send_keys('검색')
btn_search.click()
```

```python
#class : 'gLFyf'
driver = wb.Chrome()
driver.get("https://www.google.com")
btn_search = driver.find_element(By.CSS_SELECTOR,'.gLFyf')
btn_search.send_keys('검색')
btn_search.send_keys(Keys.ENTER)
```

### Keys.PAGE_DOWN 활용

```python
driver = wb.Chrome()
driver.get("https://www.google.com")
btn_search = driver.find_element(By.CSS_SELECTOR,'.gLFyf')
btn_search.send_keys('빅데이터')
btn_search.send_keys(Keys.ENTER)
```

```python
driver = wb.Chrome()
driver.get("https://www.google.com")
body = driver.find_element(By.CSS_SELECTOR,'.gLFyf')
body.send_keys('빅데이터') 
body.send_keys(Keys.ENTER)
               
for i in range(3):
    body = driver.find_element(By.TAG_NAME,'body')
    body.send_keys(Keys.PAGE_DOWN) 
    time.sleep(1)
```

```python
url='https://www.youtube.com/'
url_sub="results?search_query="
query='빅데이터'
driver = wb.Chrome()
driver.get(url+url_sub+query)

for i in range(5):
    body = driver.find_element(By.TAG_NAME,'body')
    body.send_keys(Keys.PAGE_DOWN) 
    time.sleep(1)
```
